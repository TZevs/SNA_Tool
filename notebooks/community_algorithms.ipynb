```python
import networkx as nx
import pandas as pd
```

```python
df = pd.read_csv('../data/facebook_combined.txt', sep=' ', names=['first_node', 'last_node'])
G = nx.Graph()
G.add_edges_from(df.values)
```

### Testing Different Community Detection Algorithms
- The algorithms available in the NetworkX library

```python
# Greedy Modularity
c = nx.community.greedy_modularity_communities(G)
print(len(c))
```

```python
# Asynchronous Label Propagation
c = nx.community.asyn_lpa_communities(G)
len(list(c))
```

```python
# Label Propagation
label_c = nx.community.label_propagation_communities(G)
len(list(c))
```

```python
# Fast Label Propagation
c = nx.community.fast_label_propagation_communities(G)
len(list(c))
```

```python
# Louvain
c = nx.community.louvain_communities(G)
len(list(c))
```

```python
# Louvain Partitions
c = nx.community.louvain_partitions(G)
len(list(c))
```

```python
# Leiden
c = nx.community.leiden_communities(G, backend='nx-cugraph')
len(list(c))
```
