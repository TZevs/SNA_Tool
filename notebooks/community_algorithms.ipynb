```python
coms_list = nx.community.louvain_communities(G)
node_coms = {}
comms_dict = {}

for index, com in enumerate(coms_list):
    for n in com:
        node_coms[n] = index

for index, com in enumerate(coms_list):
    comms_dict[f'c{index}'] = list(com)

#print(coms_dict['c0'][120])
```

```python
density = {}

for cid, nodes in coms_dict.items():
    sub = G.subgraph(nodes)
    density[cid] = nx.density(sub)

print(density)
```

```python
intra_deg = {}
for cid, nodes in coms_dict.items():
    com_subgraph = G.subgraph(nodes)
    sub_degree = nx.degree_centrality(com_subgraph)
    intra_deg[cid] = sub_degree

print(intra_deg)
```

```python
intra_deg = defaultdict(lambda: defaultdict(int))
inter_deg = defaultdict(lambda: defaultdict(int))

for u, v in G.edges():
    comm_u = node_coms[u]
    comm_v = node_coms[v]

    if comm_u == comm_v:
        intra_deg[comm_u][u] += 1
        intra_deg[comm_v][v] += 1
    else:
        inter_deg[comm_u][u] += 1
        inter_deg[comm_v][v] += 1

sorted_intra = {
    comm_id: dict(sorted(nodes.items(), key=lambda item: item[1], reverse=True))
    for comm_id, nodes in intra_deg.items()
}
print(sorted_intra)
```

```python
community_size = defaultdict(int)
for cid, com in coms_dict.items():
    community_size[cid] = len(com)

print(community_size)
```

```python
top_10_pc = {}

for cid, nodes in sorted_intra.items():
    node_ls = list(nodes)
    size = community_size[cid]

    top_count = math.ceil(size * 0.10)
    top_nodes = node_ls[:top_count]

    top_10_pc[cid] = top_nodes

print(type(top_10_pc))
```

```python
community_truss = {}
for cid, nodes in comms_dict.items():
    H = G.subgraph(nodes).copy()
    truss_lvls = {n: 0 for n in H.nodes()}
    k=2

    while True:
        try:
            K = nx.k_truss(H, k)
        except nx.NetworkXError:
            break

        if K.number_of_edges() == 0:
            break

        for n in K.nodes():
            truss_lvls[n] = k

        k += 1

    community_truss[cid] = truss_lvls

print(community_truss)
```

```python
community_node_core_nums = {}

for cid, nodes in coms_dict.items():
    sub = G.subgraph(nodes)
    core = nx.core_number(sub)
    community_node_core_nums[cid] = core

community_node_core_nums = {
    comm_id: dict(sorted(nodes.items(), key=lambda item: item[1], reverse=True))
    for comm_id, nodes in community_node_core_nums.items()
}
print(community_node_core_nums)
```
