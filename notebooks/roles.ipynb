```python
import pandas as pd
import numpy as np
import networkx as nx
```

```python
df = pd.read_csv('../data/raw/facebook_combined.txt', sep=' ', names=['first_node', 'last_node'])
df.head()
```

```python
G = nx.Graph()
G.add_edges_from(df.values)
```

```python
df['empty'] = [np.empty(0,dtype=str)]*len(df)
```

```python
df.head()
```

```python
df.loc[0, 'empty'] += 'sup'
df.head()
```

```python
coms_list = nx.community.louvain_communities(G)
node_coms = {}

for index, com in enumerate(coms_list):
    for n in com:
        node_coms[n] = f'c{index}'

comms_dict = {}
for index, com in enumerate(coms_list):
    comms_dict[f'c{index}'] = list(com)

print(comms_dict)
```

```python
from collections import defaultdict
```

```python
local_part = {}

for node in G.nodes():
    deg = G.degree(node)
    if deg == 0:
        local_part[node] = 0
        continue

    edge_to_comm_count = defaultdict(int)
    for ngb in G.neighbors(node):
        edge_to_comm_count[node_coms[ngb]] += 1
```
